# MATPLOTLIB LIBRARY TUTORIAL

```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
# Data for plotting

x = np.linspace(0, 10, 25)

# y  = x (squared) + 2

y = x ** 2 + 2
```

```python
x
```

```python
y
```

```python
print(np.array([x, y]).reshape(25, 2))
```

```python
plt.plot(x, y, 'r')
```

# **SUBPLOT**

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(x, y, 'b--')
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b*-')
```

Plotting a Graph on another graph

```python
fig = plt.figure()

axes1 = fig.add_axes([0.2, 0.2, 0.8, 0.8])
axes2 = fig.add_axes([0.3, 0.5, 0.3, 0.4])

axes1.plot(x, y, 'r*-')
axes2.plot(y, x, 'b--')
```
