```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = {
    'name': ['Chard', 'Odhiambo', 'Omollo', 'John', 'Peter', 'James', 'Jack', 'Silva', 'Mathew', 'Levi'],
    'subject': ['Chem', 'Physics', 'Maths', 'Kiswahili', 'English', 'Geog', 'Biology', 'Business', 'Computer', 'History'],
    'marks': [89, 93, 91, 67, 85, 92, 74, 60, 95, 81],
    'age': [23, 24, 19, 25, 27, 18, 19, 43, 32, 22]
}
```

```python
data = pd.DataFrame(data)
data.set_index('name', inplace=True)
```

```python
data.head()
```

```python
plt.figure(figsize=(10, 3))
sns.lineplot(data = data['age'], label = 'Age')
sns.lineplot(data = data['marks'], label = 'Marks')
```
