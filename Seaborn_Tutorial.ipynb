# *Flight Delay Analysis - US Transportation*

```python
flight_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/flight_delays.csv', index_col='Month')
```

```python
flight_data.head()
```

```python
flight_data.columns
```

```python
flight_data.tail()
```

```python
plt.figure(figsize=(12, 6))
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')

sns.barplot(x=flight_data.index, y=flight_data['NK'])

plt.ylabel("Arrival delay (in minutes)")
```

```python
flight_data.columns
```

```python
plt.figure(figsize=(12, 6))
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')

sns.barplot(x=flight_data.index, y=flight_data['AA'])

plt.ylabel("Arrival delay (in minutes)")
```

```python
flight_data.head()
```

```python
plt.figure(figsize=(14, 6))
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')

sns.barplot(x=flight_data.index, y=flight_data['B6'])

plt.ylabel("Arrival delay (in minutes)")
```

```python
plt.figure(figsize=(8, 6))
plt.title('Average Arrival Delay for Spirit Airlines Flights by Month')

sns.barplot(x=flight_data['B6'], y=flight_data.index)

plt.xlabel("Arrival delay (in minutes)")
```

# ***HeatMap***

```python
plt.figure(figsize = (12, 6))

sns.heatmap(data=flight_data, annot=True)

plt.xlabel('Airline')
```
